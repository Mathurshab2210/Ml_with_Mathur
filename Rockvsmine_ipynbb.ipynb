```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data=pd.read_csv('/content/Copy of sonar data.csv',header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
x=sonar_data.drop(columns=60,axis=1)
```

```python
y=sonar_data[60]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.12,stratify=y,random_state=1)#stratify to equally distribute data
```

```python
x_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model
```

```python
x_perd=model.predict(x_train)
```

```python
tr_acc=accuracy_score(x_perd,y_train)
```

```python
print('acc on train data', tr_acc*100)
```

```python
x_test_pred=model.predict(x_test)
tst_acc=accuracy_score(x_test_pred,y_test)
```

```python
print('testing score:',tst_acc)
```
