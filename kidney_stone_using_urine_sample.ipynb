```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import r2_score,classification_report
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
ksudata=pd.read_csv('/content/drive/MyDrive/ML_practtice/kindey stone urine analysis.csv')
```

```python
ksudata.head()#0-no stone 1-stone prsent
```

```python
corrmat=ksudata.corr()
f, ax = plt.subplots(figsize =(9, 8))
sns.heatmap(corrmat, ax = ax, cmap ="RdYlBu", linewidths = 0.1)
```

```python
sns.pairplot(ksudata, kind="scatter")
plt.show()
```

```python
x=ksudata.drop('target',axis=1)
y=ksudata['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.2, random_state=3,stratify=y)
```

```python
ksudata.shape
```

```python
print(x.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data : ', accuracy_on_training_data)
prediction_on_test_data = model.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data : ', accuracy_on_test_data)
y_train_pred=cross_val_predict(model,X_train,Y_train,cv=3)
cm0=confusion_matrix(Y_train,y_train_pred)
print(cm0)
print("precision LR",precision_score(Y_train,y_train_pred))
print("recall LR",recall_score(Y_train,y_train_pred))
print("f-1 score LR",f1_score(Y_train,y_train_pred))

clf2= KNeighborsClassifier()
clf2.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data KNN: ', accuracy_on_training_data)
prediction_on_test_data = clf2.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data KNN : ', accuracy_on_test_data)
y_train_pred=cross_val_predict(clf2,X_train,Y_train,cv=3)
cm1=confusion_matrix(Y_train,y_train_pred)
print(cm1)
print("precision KNN",precision_score(Y_train,y_train_pred))
print("recall KNN",recall_score(Y_train,y_train_pred))
print("f-1 score KNN",f1_score(Y_train,y_train_pred))


clf3= RandomForestClassifier()
clf3.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data Rf: ', accuracy_on_training_data)
prediction_on_test_data = clf3.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data Rf : ', accuracy_on_test_data)
y_train_pred=cross_val_predict(clf3,X_train,Y_train,cv=3)
y_train_pred=cross_val_predict(clf3,X_train,Y_train,cv=3)
cm2=confusion_matrix(Y_train,y_train_pred)
print(cm2)
print("precision Rf",precision_score(Y_train,y_train_pred))
print("recall Rf",recall_score(Y_train,y_train_pred))
print("f-1 score Rf",f1_score(Y_train,y_train_pred))

clf4= DecisionTreeClassifier()
clf4.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data DT: ', accuracy_on_training_data)
prediction_on_test_data = clf4.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data DT: ', accuracy_on_test_data)
y_train_pred=cross_val_predict(clf4,X_train,Y_train,cv=3)
y_train_pred=cross_val_predict(clf4,X_train,Y_train,cv=3)
cm3=confusion_matrix(Y_train,y_train_pred)
print(cm3)
print("precision DT",precision_score(Y_train,y_train_pred))
print("recall DT",recall_score(Y_train,y_train_pred))
print("f-1 score DT",f1_score(Y_train,y_train_pred))


clf5= SVC()
clf5.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data SVM ', accuracy_on_training_data)
prediction_on_test_data = clf5.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data SVM : ', accuracy_on_test_data)
y_train_pred=cross_val_predict(clf5,X_train,Y_train,cv=3)
y_train_pred=cross_val_predict(clf5,X_train,Y_train,cv=3)
cm4=confusion_matrix(Y_train,y_train_pred)
print(cm4)
print("precision SVC",precision_score(Y_train,y_train_pred))
print("recall SVC",recall_score(Y_train,y_train_pred))
print("f-1 score SVC",f1_score(Y_train,y_train_pred))


clf6= XGBClassifier()
clf6.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data XGB: ', accuracy_on_training_data)
prediction_on_test_data = clf6.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data XGB: ', accuracy_on_test_data)
y_train_pred=cross_val_predict(clf6,X_train,Y_train,cv=3)
y_train_pred=cross_val_predict(clf6,X_train,Y_train,cv=3)
cm5=confusion_matrix(Y_train,y_train_pred)
print(cm5)
print("precision XGB",precision_score(Y_train,y_train_pred))
print("recall XGB",recall_score(Y_train,y_train_pred))
print("f-1 score XGB",f1_score(Y_train,y_train_pred))


clf7= MultinomialNB()
clf7.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data MNB: ', accuracy_on_training_data)
prediction_on_test_data = clf7.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data MNB : ', accuracy_on_test_data)
y_train_pred=cross_val_predict(clf7,X_train,Y_train,cv=3)
y_train_pred=cross_val_predict(clf7,X_train,Y_train,cv=3)
cm6=confusion_matrix(Y_train,y_train_pred)
print(cm6)
print("precision MNB",precision_score(Y_train,y_train_pred))
print("recall MNB",recall_score(Y_train,y_train_pred))
print("f-1 score MNB",f1_score(Y_train,y_train_pred))




clf8= GaussianNB()
clf8.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data GNB: ', accuracy_on_training_data)
prediction_on_test_data = clf8.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data GNB : ', accuracy_on_test_data)
y_train_pred=cross_val_predict(clf8,X_train,Y_train,cv=3)
y_train_pred=cross_val_predict(clf8,X_train,Y_train,cv=3)
cm7=confusion_matrix(Y_train,y_train_pred)
print(cm7)
print("precision GNB",precision_score(Y_train,y_train_pred))
print("recall GNB",recall_score(Y_train,y_train_pred))
print("f-1 score GNB",f1_score(Y_train,y_train_pred))


clf9=BernoulliNB()
clf9.fit(X_train, Y_train)
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data BNB: ', accuracy_on_training_data)
prediction_on_test_data = clf9.predict(X_test)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data BNB: ', accuracy_on_test_data)
y_train_pred=cross_val_predict(clf9,X_train,Y_train,cv=3)
y_train_pred=cross_val_predict(clf9,X_train,Y_train,cv=3)
cm8=confusion_matrix(Y_train,y_train_pred)
print(cm8)
print("precision BNB",precision_score(Y_train,y_train_pred))
print("recall BNB",recall_score(Y_train,y_train_pred))
print("f-1 score BNB",f1_score(Y_train,y_train_pred))








```
